```python
import requests
from bs4 import BeautifulSoup

res = requests.get('https://comic.naver.com/webtoon/weekday.nhn')

html = res.text
soup = BeautifulSoup(html, 'html.parser')

titles = soup.select('#content > div.list_area.daily_all > div > div > ul > li > a')
webtoons_date = {}
current_webtoon_list = []
for title in titles:
    name = title.text
    date = title['href'].split('=')[-1]
    webtoons_date[name] = date
    current_webtoon_list.append(name)
    
print(webtoons_date)
```

```python
import pandas as pd

webtoon_data = pd.Series(webtoons_date)
webtoons = pd.DataFrame({'date' : webtoon_data})
webtoons
```

```python
res2 = requests.get('https://comic.naver.com/webtoon/genre.nhn')

html2 = res2.text
soup2 = BeautifulSoup(html2, 'html.parser')

genres = soup2.select('#content > div.snb > ul > li')

genre_list = []
for genre in genres:
    genre_list.append(genre.a['href'].split('=')[-1])
    
print(genre_list)
```

```python
res3 = requests.get('https://comic.naver.com/webtoon/genre.nhn?genre=episode')

html3 = res3.text
soup3 = BeautifulSoup(html3, 'html.parser')

enames = soup3.select('#content > div.list_area > ul > li')
ename_list = []
for ename in enames:
    ename_list.append(ename.a['title'])
    
print(ename_list)
```

```python
episode_webtoon = {}
for current_webtoon in current_webtoon_list:
    if current_webtoon in ename_list:
        episode_webtoon[current_webtoon] = 'episode'
        
print(episode_webtoon)
```

```python
webtoon_genre = {}
for genre in genre_list:
    res4 = requests.get('https://comic.naver.com/webtoon/genre.nhn?genre=' + genre)

    html4 = res4.text
    soup4 = BeautifulSoup(html4, 'html.parser')

    genre_names = soup4.select('#content > div.list_area > ul > li')
    genre_name_list = []
    for genre_name in genre_names:
        genre_name_list.append(genre_name.a['title'])
        
       
        for current_webtoon in current_webtoon_list:
            if current_webtoon in genre_name_list:
                webtoon_genre[current_webtoon] = genre

print(webtoon_genre)
```

```python
webtoon_dates = pd.Series(webtoons_date)
webtoon_genres = pd.Series(webtoon_genre)
webtoons = pd.DataFrame({'date' : webtoon_dates, 
                        'genre' : webtoon_genres})
webtoons
```

```python
res6 = requests.get('https://comic.naver.com/webtoon/genre.nhn?genre=episode')

html6 = res6.text
soup6 = BeautifulSoup(html6, 'html.parser')

arts2 = soup6.select('#content > div.list_area > ul > li')
rating_stars_list = []
for art in arts2:
    rating_stars = art.find('div', class_ = 'rating_type').find('strong').text
    rating_stars_list.append(rating_stars)
print(rating_stars_list)
```

```python
webtoon_cartoonist = {}
webtoon_latest_update = {}
webtoon_average_rating_star = {}
for genre in genre_list:

    res5 = requests.get('https://comic.naver.com/webtoon/genre.nhn?genre=' + genre)

    html5 = res5.text
    soup5 = BeautifulSoup(html5, 'html.parser')

    arts = soup5.select('#content > div.list_area > ul > li')
    arts_dict = {}
    for art in arts:
        cartoonist = art.dd.a.text
        name = art.a['title']
        arts_dict[name] = cartoonist
        for current_webtoon in current_webtoon_list:
            if current_webtoon in arts_dict.keys():
                webtoon_cartoonist[current_webtoon] = arts_dict[current_webtoon]
                
    for art in arts:
        latest_update = art.find('dd', class_ = 'date2').text
        name = art.a['title']
        arts_dict[name] = latest_update
        for current_webtoon in current_webtoon_list:
            if current_webtoon in arts_dict.keys():
                webtoon_latest_update[current_webtoon] = arts_dict[current_webtoon]    
    
    
    for art in arts:
        average_rating_stars = art.find('div', class_ = 'rating_type').find('strong').text
        name = art.a['title']
        arts_dict[name] = average_rating_stars
        for current_webtoon in current_webtoon_list:
            if current_webtoon in arts_dict.keys():
                webtoon_average_rating_star[current_webtoon] = arts_dict[current_webtoon]  

print(webtoon_average_rating_star)
```

```python
webtoon_dates = pd.Series(webtoons_date)
webtoon_genres = pd.Series(webtoon_genre)
webtoon_cartoonists = pd.Series(webtoon_cartoonist)
webtoon_latest_updates = pd.Series(webtoon_latest_update)
webtoon_average_rating_stars = pd.Series(webtoon_average_rating_star)
webtoons4 = pd.DataFrame({'date' : webtoon_dates, 
                        'genre' : webtoon_genres,
                         'cartoonist' : webtoon_cartoonist,
                         'latest update' : webtoon_latest_updates,
                         'average rating star' : webtoon_average_rating_stars})
webtoons4
```
