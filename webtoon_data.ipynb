```python
import requests
from bs4 import BeautifulSoup
import operator

res = requests.get('https://comic.naver.com/webtoon/weekday.nhn')

html = res.text
soup = BeautifulSoup(html, 'html.parser')

titles = soup.select('#content > div.list_area.daily_all > div > div > ul > li > a')
webtoon_lists = []
date_lists = []
webtoons_info2 = {}
for title in titles:
    name = title.text
    date = title['href'].split('=')[-1]
    webtoons_info2[name] = date
    
print(webtoons_info2)

import pandas as pd

webtoon_data = pd.Series(webtoons_info2)
webtoons = pd.DataFrame(webtoon_data, columns = ['date'])
print(webtoons)
```
